```python
import lectura
```

```python
train_data = lectura.load_data('datasets/train_en_A.tsv')
dev_data = lectura.load_data('datasets/dev_en_A.tsv')

train_data['text_processed'] = train_data['text'].apply(lectura.english_tokenizer)
dev_data['text_processed'] = dev_data['text'].apply(lectura.english_tokenizer)
```

```python
y_train = train_data['HS'].values
y_dev = dev_data['HS'].values
```

```python
from sklearn.feature_extraction.text import CountVectorizer

ngrams = CountVectorizer(ngram_range=(1,1))
X_train_ngrams = ngrams.fit_transform(train_data['text_processed'].apply(' '.join))
X_dev_ngrams = ngrams.transform(dev_data['text_processed'].apply(' '.join))
```

```python
from sklearn import svm, linear_model, tree, ensemble
import numpy as np

classifiers = {'logistic_regression': 
                   {'model': linear_model.LogisticRegression(),
                    'params': {'clf__penalty': ['l1','l2'],
                               'clf__tol': [0.0001,0.001,0.01,0.1],
                               'clf__C': np.logspace(-4, 2)}
                   }, 
               'linear_SVC': 
                   {'model': svm.LinearSVC(dual=False),
                    'params': {'clf__penalty': ['l1','l2'],
                               'clf__tol': [0.0001,0.001,0.01,0.1],
                               'clf__C': np.logspace(-4, 2)}}, 
               'random_forest': 
                   {'model': ensemble.RandomForestClassifier(),
                    'params': {'clf__n_estimators': np.arange(10, 110, 10),
                               'clf__max_depth': np.append(np.arange(1, 11), None)}} 
}
```

```python
import evaluation

evaluation.evaluate(X_train_ngrams, y_train, X_dev_ngrams, y_dev, classifiers, False)
```
